```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import matplotlib.pyplot as pyplot
import seaborn; seaborn.set()
```

```python
Riders_file = open('riders.txt', 'r')

Riders = sorted(Riders_file.readlines())
print(Riders)
```

```python
dataframe_collection = {}  #dictionary collecting all names
cols = ["Year", "Race", "Position", "Time", "Gap", "Elevation", "Distance" ,"Over 2000", "Over 1800",]

for rider in Riders:
    new_data = np.random.rand(3,len(cols))
    dataframe_collection[rider] = pd.DataFrame(new_data, columns=cols)

for key in dataframe_collection.keys():
    print("\n" +"="*40)
    print(key)
    print("-"*40)
    print(dataframe_collection[key])
```
