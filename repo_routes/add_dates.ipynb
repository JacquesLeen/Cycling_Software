```python
import pandas as pd
```

```python
results = pd.read_csv(r'C:\Users\yd26114\Desktop\fc\Tables\All_Races_Results_Info.csv', dtype={'Race Code':'string', 'Year': 'string'})
race_generic_info = pd.read_csv(r'C:\Users\yd26114\Desktop\fc\Tables\All_Races_Generic_Info.csv', dtype={'Race Code':'string', 'Year': 'string'})
```

```python
race_generic_info.drop(['Unnamed: 0'], axis =1, inplace=True)
results.drop(['Unnamed: 0'], axis =1, inplace=True)
```

```python

test_race_generic_info = race_generic_info.drop_duplicates(subset = race_generic_info.columns.difference(['Month']))
```

```python
results['Full_Race_Code'] = results['Race Code']+ results['Year']
test_race_generic_info['Full_Race_Code'] = test_race_generic_info['Race Code'] + test_race_generic_info['Year']

display(results.head())
display(test_race_generic_info.head())
```

```python
for_merge = test_race_generic_info[['Date', 'Full_Race_Code']]
display(for_merge)
```

```python
results= results.merge(for_merge, how='left', on='Full_Race_Code')

display(results)
```
